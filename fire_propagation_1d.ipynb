```python
nx = 100
nt = 100
t0 = 0; tf = 5; 
x0, xf = 0, 100
dt = ( tf - t0 )/nt
dx = ( xf - x0 )/nx

U = np.zeros( (nt, nx) )
U[0, 50] = 1
D = 5
k = 2

for j in range(0, nt-1):
    for i in range(1, nx-1):
        U[j+1, i] = D * dt / dx * (U[j, i-1] - 2*U[j, i] + U[j, i+1]) +  k * U[j, i] * (1 - U[j, i]) * dx*dx * dt + U[j, i]

plt.imshow(U)
```
